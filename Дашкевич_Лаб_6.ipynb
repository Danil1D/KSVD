```python

!pip install kagglehub ultralytics

import kagglehub
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt

dataset_path = kagglehub.dataset_download('shlokraval/ppe-dataset-yolov8')

print("–§–∞–π–ª–∏ —É –¥–∞—Ç–∞—Å–µ—Ç—ñ:")
print(os.listdir(dataset_path))

model = YOLO('yolov8n.pt')
yaml_path = os.path.join(dataset_path, 'data.yaml')

model.train(
    data=yaml_path,
    epochs=5,
    imgsz=320,
    batch=32,
    device=0,
    project='ppe_detection_colab',
    name='yolov8n_experiment'
)

results_path = '/content/ppe_detection_colab/yolov8n_experiment/results.png'

if os.path.exists(results_path):
    img = plt.imread(results_path)
    plt.figure(figsize=(10,6))
    plt.imshow(img)
    plt.axis('off')
    plt.title('–ö—Ä–∏–≤—ñ –Ω–∞–≤—á–∞–Ω–Ω—è (Loss, mAP, Precision, Recall)')
    plt.show()
else:
    print("–§–∞–π–ª results.png –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ.")

metrics = model.val()

print("\nüìä –û—Ü—ñ–Ω–∫–∞ –º–æ–¥–µ–ª—ñ:")
print(f"Mean Precision: {metrics.box.p.mean():.4f}")
print(f"Mean Recall: {metrics.box.r.mean():.4f}")
print(f"mAP@0.5: {metrics.box.map50:.4f}")
print(f"mAP@0.5-0.95: {metrics.box.map:.4f}")


model.predict(
    source=os.path.join(dataset_path, 'valid', 'images'),
    save=True,
    imgsz=320,
    conf=0.25,
    stream=True
)
```
